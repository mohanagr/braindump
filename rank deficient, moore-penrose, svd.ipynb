```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
x = np.linspace(-4,5,8)
A = np.vander(x,N=10,increasing=True)
print(A.shape)
m_true = np.random.randn(10).reshape(10,-1)
y = A@m_true
y_obs = y + np.random.randn(len(y)).reshape(len(y),-1)
```

```python
plt.plot(y_obs)
```

```python
H = A.T@A
# print(H)
H_inv = np.linalg.pinv(H)
m = H_inv@A.T@y_obs
```

```python
y_pred = A@m
plt.plot(y_pred-y)
```
