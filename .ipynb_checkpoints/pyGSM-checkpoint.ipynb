```python
from pygdsm import GlobalSkyModel2016
from pygdsm import LowFrequencySkyModel
import healpy as hp
import numpy as np
```

```python
gsm_2016 = GlobalSkyModel2016(freq_unit='MHz',resolution='hi')
map1 = gsm_2016.generate(10)
```

```python
hp.mollview(np.log2(map1),min=16,max=20,title="10 MHz with 1Â° beam",unit='log Temp')
```

```python
map2=hp.sphtfunc.smoothing(map1,fwhm=7.46*np.pi/180)
```

```python
hp.mollview(np.log2(map2),min=16,max=20, title="10 MHz with 7.5Â° beam",unit='log Temp')
```

```python
lwa = LowFrequencySkyModel(freq_unit='MHz')
x=lwa.generate(10)
lwa.view(logged=True)
```

```python
hp.get_nside(x)
```

```python
hp.nside2resol(2)*180/np.pi
```

```python
y=hp.ud_grade(x,)
```

```python
hp.mollview(np.log2(y),min=16,max=20)
```

```python
z1 = hp.ud_grade(hp.sphtfunc.smoothing(y,sigma=2*np.pi/180),64)
```

```python
hp.mollview(np.log2(z1),min=16,max=20)
```
