```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,2*np.pi,1001)
```

```python
x=np.cos(t)
y=np.sin(t)
f=plt.gcf()
f.set_size_inches(4,4)
plt.plot(x,y)
```

```python
lm = np.diag((1/4,1/9))
theta=np.deg2rad(0)
c, s = np.cos(theta), np.sin(theta)
R = np.array(((c, -s), (s, c)))
```

```python
unrot = np.vstack((x,y))
```

```python
rot = R@lm@unrot
```

```python
f=plt.gcf()
# f.set_size_inches(4,4)
plt.plot(rot[0,:],rot[1,:])
```
