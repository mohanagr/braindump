```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
x=np.linspace(0,5,1001)
ph=x**2 - 2.5*np.exp(-0.5*x**2)+2.5
plt.plot(x,ph)
```

```python
def get_ts(nu,N,fs,nchan):
    bw=fs/(2*nchan)
    n=np.arange(0,N)
    ts = np.sinc(bw*n/fs)*np.sin(2*np.pi*nu*n/fs)
    return ts + 1*np.random.randn(N)
```

```python
fs=250e6
nchan=2048
N=nchan *40
```

```python
ts1 = get_ts(138e6,N,fs,nchan)
```

```python
plt.plot(ts1[:100],'--')
```

```python
tau = ph/(2*np.pi*138e6)
```

```python
def interp_ts(tau,dt_tau,ts,dt_ts):
    maxt = dt_ts*ts.shape[0]
    nn = int(maxt/dt_tau)
    taunew = tau[:nn].copy()
    tau_t = np.arange(0,nn)*dt_tau
    ts_t = np.arange(0,ts.shape[0])*dt_ts
    taunew_interp = np.interp(ts_t,tau_t,taunew)
    ts_new = np.interp(ts_t+taunew_interp,ts_t,ts)
    return ts_new
```
