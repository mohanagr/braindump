```python
from astropy.cosmology import Planck18 as p18
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(1,100,1001)
c = 3e5 #km/s
nu0 = 10e6 #10 MHz
nu21 = 1.42e9 #1420 MHz
res = 61e3 #61 kHz
eta = 1/res
```

```python
Dc = p18.comoving_distance(z)
E = p18.efunc(z)
```

```python
k_t = 2*np.pi*nu0*15/(c*Dc)

k_p = 2*np.pi*nu21*100*E*eta/(c*(1+z)**2)
```

```python
plt.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
```

```python
f=plt.gcf()
f.set_size_inches(5,4)
plt.figure(dpi=500)
plt.loglog(z,k_t/p18.h, label=r"$k_{\perp}$ for 15 km baseline")
plt.loglog(z,k_p, label=r"$k_{\parallel}$ for 61 kHz resolution")
plt.ylabel(r"$k_{\parallel}\;\; or\;\; k_{\perp} [hMpc^{-1}]$",fontsize=14)
plt.xlabel(r"Redshift z", fontsize=14)
plt.legend()
```
