```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
lat=np.deg2rad(90)
decl = np.deg2rad(90)
phi = np.linspace(0,2*np.pi,1001)
alpha = (4/24)*2*np.pi
```

```python
#start with phi = 0, from RA = 0, with just one east west baseline.
be=1
bn=0
u = be*np.cos(phi)*np.cos(decl) - bn*np.sin(lat)*np.cos(decl)*np.sin(phi)
v = -be*np.sin(decl)*np.sin(phi) + bn*np.sin(lat)*np.sin(decl)*np.cos(phi) + bn*np.cos(lat)*np.cos(decl)
```

```python
plt.scatter(u,v)
plt.xlim(-1,1)
plt.ylim(-1,1)
```
