```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
sos = signal.butter(12, 0.01, 'low',output='sos')
x = signal.unit_impulse(8192)
y_sos = signal.sosfilt(sos, x)
```

```python
sos = signal.ellip(12, 0.1, 80, 0.06, btype='low', analog=False, output='sos')
x = signal.unit_impulse(8192)
y_sos = signal.sosfilt(sos, x)
```

```python
plt.plot(y_sos)
# plt.xlim(2000,2010)
# plt.ylim(-5e-6,1e-6)
```

```python
w,h=signal.sosfreqz(sos,8192) #uses RFFT that's 8192 long
```

```python
plt.plot(w/(np.pi),np.abs(h))
plt.axvline(0.03)
plt.xlabel('fraction of Nyquist freq')
plt.semilogy()
plt.xlim(0,0.5)
plt.ylim(1e-10,10)
```

```python
plt.plot(w/(np.pi),np.abs(h)**2)
plt.axvline(0.01,ls='-.')
plt.xlabel('fraction of Nyquist freq')
plt.semilogy()
plt.xlim(0,0.1)
plt.ylim(1e-10,10)
```
