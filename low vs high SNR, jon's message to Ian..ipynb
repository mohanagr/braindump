```python
import numpy as np
import matplotlib.pyplot as plt



# Generate x values
x = np.linspace(-10, 10, 10000)  # use a dense range of x values for a smooth curve
y = 2*np.sinc(x)

# Plot the sinc function


# Specify the y-value range for shading
a = 1

f = plt.gcf()
f.set_size_inches(10,4)
plt.suptitle("let noise RMS = 1")
plt.subplot(121)
plt.plot(x, y, label='sinc(x)')
# Shade the area where the y-value of the sinc function is between -a and a
plt.fill_between(x, -a, a, where=(y <= a) & (y >= -a), color='orange', alpha=0.5, label='Area between -a and a')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('SNR = 2')
plt.axhline(0,ls='--',c='black')
plt.xlim(-1.5,1.5)

# Generate x values
x = np.linspace(-10, 10, 10000)  # use a dense range of x values for a smooth curve
y = 10*np.sinc(x)

plt.subplot(122)
plt.plot(x, y, label='sinc(x)')
# Shade the area where the y-value of the sinc function is between -a and a
plt.fill_between(x, -a, a, where=(y <= a) & (y >= -a), color='orange', alpha=0.5, label='Area between -a and a')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('SNR = 10')
plt.axhline(0,ls='--',c='black')
plt.xlim(-1.5,1.5)

# Display the plot
plt.show()
```

just to visually illustrate what Jon's said here. in order to beat the noise, we need independent samples. we will get independent samples every 1/bandwidth (zero crossings), RMS around zero crossing is the typical noise level.
