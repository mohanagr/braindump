```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_ts (p , k_center , dk , N ):
    k = k_center + dk
    return np . cos (2* np . pi * np . arange (0 , N )* k / N + 2* np . pi * dk * p )
```

```python
ph =[]
ft = []
dk = 0.1 # < |0.5| , beyond 0.5 gets into next channel
N = 16384
k_center = 20 # can be anything < N /2
for p in range (0 ,10):
    f = np . fft . rfft ( get_ts (p , k_center , dk , N ))
    ph . append ( np . angle ( f [ k_center ]))
    ft.append(f[k_center])
    plt.plot(np.abs(f))
    plt.xlim(0,40)
m , c = np . polyfit ( np . arange (0 ,10) , np . unwrap ( ph ) ,1)
print ( m /(2* np . pi )) # = dk
# plt.plot(np.abs(ft))
# plt.xlim(0,10)
```
