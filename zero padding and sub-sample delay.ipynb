```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(-2,2,101)
# y = np.random.randn(len(x))
y = np.exp(-x**2/10)
# plt.plot(x,y)
y2 = np.hstack([y,np.zeros(len(y))])
plt.plot(y2)
```

```python
f=np.fft.rfft(y2)
f[:] = f*np.exp(-2j*np.pi*np.arange(len(f))*1.5/len(y2))
ynew=np.fft.irfft(f)
```

```python
plt.plot(ynew)
```
