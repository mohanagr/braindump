```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def quantize_nbits(x,nbits=18,high=None,low=None):
    #old version
#     print("received nbits", nbits)
    nlev = (1<<nbits)-1
#     print(nlev, "received levels")
    if(not high):
        high=x.max()
        low=x.min()
    bitmax = (nlev-1)//2
    return np.round(2*bitmax*x/(high-low) - bitmax * (high+low)/(high-low)).astype(int)


def quantize(x,nbits=4,delta=0.1):
    nlev=(1<<nbits)-1
    bitmax=(nlev-1)//2
    print("bitmax is", bitmax)
    print()
#     print(np.round(x/delta))
    return np.clip(np.round(x/delta),-bitmax,bitmax)
```

```python
fs=10000
f=10
t=1
N=t*fs
n=np.arange(0,N)
x=np.sin(2*np.pi*f*n/fs)
```

```python
plt.plot(x)
```

```python
qx=quantize(1.8*x,delta=0.3)
```

```python
plt.plot(qx)
```

Smaller delta = better resolution and low quantization error

but it also means you'll saturate more easily for a fixed number of levels.
