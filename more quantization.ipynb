```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sig_inf(sigma,n,rho):
    N=1000000
    x = (2*sigma**2*n**2 + n**4 + (1+2*rho**2)*sigma**4)/N + (1-1/N)*sigma**4*rho**2 - rho**2*sigma**4
    return np.sqrt(x)

def sig_2(sigma,n,rho):
    N = 1000000
    rho2 = 2/np.pi * np.arctan(sigma**2*rho/np.sqrt(2*sigma**2*n**2 + n**4 + sigma**4*(1-rho**2)))
#     print(rho2)
    x = 1/N + (1-1/N)*rho2**2 - rho2**2
    return np.sqrt(x)
```

```python
snr = np.linspace(0.0001,10,16000)
rho_corr_coef=1
rho_V = rho_corr_coef * snr/(1+snr)
rho_2 = 2/np.pi * np.arcsin(rho_V)
# print(rho_2)
tot_power = 1 + snr
```

```python
rho_y = np.linspace(0,0.99,1001)
plt.plot(rho_y, 2/(np.pi*np.sqrt(1-rho_y**2)))
```

```python
err_r2 = tot_power*np.sqrt(1-rho_V**2)*((np.pi/2)**2 - np.arcsin(rho_V)**2)**0.5/1000
```

```python
sigma_V = (1/(1-rho_V)**2) *np.pi/2 * np.cos(np.pi/2 * rho_2) * err_r2 #sig_2(np.sqrt(tot_power),0,rho_V)
sigma_inf = sig_inf(np.sqrt(tot_power),0,rho_V)
```

```python
fig = plt.gcf()
fig.set_size_inches(7.5,4)
fig.set_facecolor('white')
plt.plot(snr,sigma_inf/sigma_V,c='black')
plt.ylim(0,1)
plt.xlim(0,10)
plt.axhline(0.32,ls='--',c='gray')
plt.axhline(0.64,ls='--',c='gray')
# plt.axhline(0.2)
# plt.axvline(2)
plt.title("x-axis scaled by 0.4")
plt.ylabel(r"Corr. efficiency ($\epsilon$)")
plt.xlabel("signal-to-noise ratio")
# plt.plot(snr,snr/sigma_inf)
```
