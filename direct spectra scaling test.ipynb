```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
pol00 = [2.0740e10, 3.5154e10, 1.4397e11, 5.2952e11, 1.5928e12, 4.9235e12, 1.5642e13, 4.967e13]
adc0 = [-0.53, -0.35, 0.4, 1.264, 2.022, 2.818, 3.653, 4.485]
input_power = np.arange(-30,10,5, dtype="float64")
print(input_power)
```

```python
f=plt.gcf()
f.set_facecolor("white")
plt.title("SNAP power scaling")
plt.plot(input_power, np.log10(pol00), marker='*', label='direct spectra')
ax=plt.gca()
ax.set_ylabel("log(pol00)")
ax.set_xlabel("input power (dBm)")
ax.text(-30, 13, "power in dbm = [log(pol00) - 13.208] / 0.1035 ")
ax.text(-30, 12.8, "adc bits = power in dbm * 0.1612 + 3.6535")
ax2=ax.twinx()
ax2.plot(input_power, adc0, marker='o',c='r', label='adc bits')
ax2.set_ylabel("RMS adc bits")
```

```python
plt.plot(np.log(pol00), adc0)
```

```python
plt.plot(input_power, adc0, marker='*')
```

```python
m,c=np.polyfit(input_power[1:], adc0[1:], 1)
print(m,c)
```

```python
m,c=np.polyfit(input_power[1:], np.log10(pol00[1:]), 1)
print(m,c)
```

```python
plt.plot(input_power, np.log10(pol00), marker='*')
plt.plot(input_power, m*input_power+c, marker='*')
```
