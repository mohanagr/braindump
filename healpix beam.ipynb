```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
from astropy.time import Time
from astropy.coordinates import SkyCoord,EarthLocation, AltAz
from astropy import units as u
```

```python
NPIX=hp.nside2npix(64)
print(NPIX)
phi,tht=hp.pix2ang(64,np.arange(NPIX),lonlat=True)
```

```python
def get_beam(alt):
    rho = (90-alt)*np.pi/180
    beam=10*np.exp(-rho**2)
    return beam
```

```python
c=SkyCoord(ra=phi*u.degree, dec=tht*u.degree)
# bear_mountain = EarthLocation(lat=45.5017*u.deg, lon=-73.5673*u.deg, height=233*u.m)
mylocation = EarthLocation(lat=-46.8964*u.deg, lon=37.7509*u.deg, height=10*u.m)
utcoffset = -4*u.hour  # Eastern Daylight Time
time = Time('2022-05-30 20:00:00') - utcoffset
z=c.transform_to(AltAz(obstime=time,location=mylocation))
```

```python
beam=get_beam(z.alt.value)
```

```python
hp.mollview(beam,title='Marion Island 23:00',cmap='inferno')
```

```python
time=Time('2022-05-30 20:00:00') - utcoffset
for i in range(10):
    plt.clf()
    time = time + 1*u.hour
    print(time+utcoffset)
    z=c.transform_to(AltAz(obstime=time,location=mylocation))
    beam=get_beam(z.alt.value)
    hp.mollview(beam,title='Marion Island 23:00')
    plt.pause(0.1)
```
