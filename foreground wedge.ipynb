```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
nu=np.linspace(10,125,2049)*1e6
l=0.01 #rad
D=1 #m dish diameter
c=3e8
b=100 #m
V=np.cos(2*np.pi*b*l*nu/c)
u=b*nu/c
du=1
umax=100
Nu=int(umax/du)
gridded=np.zeros(Nu)
```

```python
plt.plot(u,V)
```

```python
for i in range(len(V)):
    idx=np.round(u[i]/du).astype(int)
#     print(idx)
    gridded[idx]+=V[i]
```
