```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def quantize(x):
    ii=x.real >=0
    jj=x.imag>=0
    return 2*ii-1 + 1j*(2*jj-1)
```

```python
mult=1000
n=np.arange(0,4096*mult)
x=np.sinc(0.2*(n-len(n)/2)/4096)*np.cos(2*np.pi*100.2*n/4096)
# plt.plot(x)
```

```python
spec=np.fft.rfft(x.reshape(-1,4096),axis=1)
```

```python
plt.imshow(np.abs(spec),aspect='auto')
```

```python
plt.plot(np.real(spec[:,100]))
```

```python
specq=quantize(spec)
```

```python
spec2=np.zeros(spec.shape,dtype='complex128')
spec2[:,100]=specq[:,100]
```

```python
x2=np.fft.irfft(spec2,axis=1).ravel()
```

```python
plt.plot(x2[5*4096-20:5*4096+20])
```

```python
spec3=np.fft.rfft(x2.reshape(-1,409600),axis=1)
```

```python
plt.imshow(np.abs(spec3[:,95*100:102*100]),aspect='auto',vmax=4)
plt.colorbar()
```
