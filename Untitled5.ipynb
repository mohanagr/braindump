```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nn=32
np.random.seed(42)
f=np.zeros(nn,dtype='complex128')
dk=2
f[:dk+1]=np.random.randn(dk+1) + 1j*np.random.randn(dk+1)
f[-dk:]=np.random.randn(dk) + 1j*np.random.randn(dk)
y=np.fft.ifft(f)
```

```python
plt.plot(y.real)
plt.plot(y.imag)
```

```python
cs=np.cos(2*np.pi*20*np.arange(nn)/nn)
si=np.sin(2*np.pi*20*np.arange(nn)/nn)
```

```python
f2=np.fft.fft(y.real*cs+y.imag*si)
```

```python
f2[20-3:20+4]*2
```

```python
y2=np.fft.ifft(f2)
```

```python
plt.plot(y2.real)
plt.plot(y2.imag)
```

```python
y2=y2.real
```

```python
#we know y2 is purely real
f2=np.fft.rfft(y2)
f3=np.hstack([f2,np.zeros(256,dtype=f2.dtype)])
y3=np.fft.irfft(f3)
```
