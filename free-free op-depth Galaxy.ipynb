```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_alpha_nu(T,nu,ne):
    # T in K
    # nu in MHz
    # ne in cm^-3
    return 0.65 * (T/1e4)**(-1.35) * nu**(-2.1) * ne**2 # pc^-1
```

```python
#distance to tau=1
nu = np.linspace(-3,2,10001)
nu = 10**nu
d_tau1 = 1/get_alpha_nu(8000, nu, 0.1)
d_tau2 = 1/get_alpha_nu(8000, nu, 0.01)
```

```python
# plt.style.use('seaborn-v0_8-darkgrid')
# plt.style.use('seaborn-v0_8-colorblind')
# plt.style.use('seaborn-v0_8-dark')
plt.style.use('dark_background')
plt.title("Distance to free-free optical depth = 1 (T=8000 K)",fontsize=14)
plt.plot(nu,d_tau1/1000,label=r"$n_e = 0.1\; cm^{-3}$",lw=3)
plt.plot(nu,d_tau2/1000,label=r"$n_e = 0.01\; cm^{-3}$",lw=3)
plt.axhline(20,ls="--",label="edge of Milky Way",lw=3)
plt.fill_between(nu[7398:8139],0,20,alpha=0.5, label="From Earth",color="lightgray")
plt.xlabel(r"$\nu$ (MHz)")
plt.ylabel(r"$D(\tau=1)$ (kpc)")
plt.ylim(0,30)
plt.xlim(0,20)
plt.legend(fontsize=10)
```
