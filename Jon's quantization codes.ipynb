```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
smin=0.01
smax=10
N=1
S=np.exp(np.linspace(np.log(smin),np.log(smax),1000))
print(S)
x=2.0/np.pi*np.arcsin(S/(S+N))
sigma_x=np.sqrt(1-x**2)
sigma_rho=np.pi/2*np.cos(np.pi/2*x)*sigma_x
sigma_s=(S+N)**2/N*sigma_rho
ref=S/(S+N)
corr_eff=(S+N)/sigma_s
```

```python
f=plt.gcf()
f.set_size_inches(8,4)
plt.plot(S,corr_eff)
```

```python
dr = sigma_rho *(S+N) + 2*S*(S+N)
ds = (S+N)*(1+(S/(S+N))**2)**0.5
```

```python
f=plt.gcf()
f.set_size_inches(8,4)
plt.axhline(0.64,ls='--',c='black')
plt.axhline(0.32,ls='--',c='black')
plt.ylim(0,1)
plt.xlim(0,10)
plt.plot(S/N,ds/dr)
```
